### **LISTAS**
CORCHETES [ ]

*Podemos hacer que el tamaño incremente o decrezca, Almacena otros tipos de variables y estructuras*

*Las listas son conjuntos ordenados. Se definen usando corchetes. Separados por coma se ponen los elementos.*

*Éstos pueden ser cualquier tipo de datos: str, int, float e incluso otras listas!*

### **Sets** 

*colecciones no ordenadas, no indexadas, mutables, que no permiten elementos duplicados.*

```python
set1 = {1,2,3,4,"la cartilla",1,2,10}
print(set1)
set1.add("python")
print(set1)
set1.update(["a","b",True, False])
print(set1)
set1.discard("la cartilla")
print(set1)
```

### **Diccionario** 

*Son mutables, y sus variables modificables a diferencia de las listas y tuplas, sus datos estan guardados por llaves, no por indices* 

```python
dicc = {"a":4, "b":5, "c":10};  
resp = "a" in dicc ;     
resp2 = dicc.get("a") ;  
resp3 = dicc.get("z","La llave no existe")

print(dicc)
print(resp)   #Devuelve un valor booleano
print(resp2)
print(resp3)
# El segundo componente es lo que devuelve en caso no exita el primero en el diccionario
# puede ser un entero, booleano 
```

```python
#LLAVES, VALORES E ITEMS
diccionario = { "a":1 , "b":2 , "c":3 , "d":4 , "e":5 , "f":6 , "g":7 , "h":8 }
llaves = diccionario.keys()
llaves2 = tuple(llaves)
print(llaves); 
print(llaves2,"\n")

valores = diccionario.values()
valores2 = tuple(valores)
print(valores); 
print(valores2,"\n")

result = diccionario.items()
result2 = list(result)
print(result); 
print(result2)
```
