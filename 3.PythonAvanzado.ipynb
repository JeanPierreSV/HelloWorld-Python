## **Metodos**

```python
# ---------------------------
# METODOS DE INSTANCIAS Y ESTATICOS
# --------------------------
class triangulo(): 
    def area(self): #Metodos de instancia: funciones que contienen la palabra reservada self
                    #Un objeto puede manda a llamarlo
        return (self.base * self.altura)/2
forma = triangulo()   # Creo mi instanciau objeto
triangulo.base = 10   # Coloco sus atributos
triangulo.altura = 20
print(forma.area())
```

```python
class triangulo(): 
    #Si podemos usar variables de clase
    numero = 2
    @staticmethod
    def area(base, altura): #Metodos que estan dentro de la clase y no necesitan "self"
        return (base * altura) / triangulo.numero
print(triangulo.area(10,20))   # Al ser un metodo estatico no se necesita una instancia
```

```python
# -------------------------------------
# METODOS DE CLASE
# -------------------------------------
# Lo usaremos cuando necesitemos usar variables de clase
class circulo():
    pi = 3.14159
    @classmethod # Los metodos de clase deben recibir un parametro obligatoriamente
    def area(cls, radio):  # CLS por convencion
        return cls.pi * radio**2
print(circulo.area(10))
```

## **Herencias**

```python
# Nos permitira crear clases a partir de otras clases ya existetes
# pero con algunas diferencias
class Animal:
    def comer(self):
        print("comiendo")
    def dormir(self):
        print("durmiendo")

class Mascota:
    def fecha_adopcion(self,fecha):
        self.fecha_de_adopcion = fecha

class Perro(Animal,Mascota):   #Herencia multiple
    def __init__(self,nombre):
        self.nombre = nombre
    def ladrar(self):
        print("ladrando")
    def dormir(self):
        print(self.nombre + " esta durmiendo")
    

class Gato(Animal): #La clase gato hereda los metodos de Animal
    def ronroneo(self):
        print("ronroneando")

can = Perro("Firulais")
can.ladrar()
can.comer()
can.dormir()   
# Primero lo buscara en la clase principal (can -> Perro) 
# luego en sus herencias (Animal,Mascota)
can.fecha_adopcion("Hoy")
print(can.fecha_de_adopcion)

bola_de_nieve = Gato()
bola_de_nieve.comer()
bola_de_nieve.dormir()
```
