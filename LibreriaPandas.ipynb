    - La imputación es el proceso de reemplazar los valores perdidos por otros valores predichos u observados. Existen diferentes tipos de imputación, pero la más común es la imputación media.

* Cual es la diferencia entre **sort_values** () function y **sort_index**() function?
    - sort_values() sorts by columns while sort_index() sorts on either columns or rows.

* ¿Cuáles de las siguientes afirmaciones son verdaderas sobre el **slicing**?
    - The .loc() function gets rows (or columns) with particular labels from the index.
    - The .iloc() function gets rows (or columns) at particular positions in the index.

    - Inner join.

## **join**

```
df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
df
```

```
other = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
other
```

```
df.join(other, lsuffix='_caller', rsuffix='_other')
```

```
df.set_index('key').join(other.set_index('key'))
```

```
df.join(other.set_index('key'), on='key')
```

## **concat**

```
s1 = pd.Series(['a', 'b'])
s2 = pd.Series(['c', 'd'])
pd.concat([s1, s2])
```

```
pd.concat([s1, s2], ignore_index=True)
```

```
pd.concat([s1, s2], keys=['s1', 's2'])
```

## **merge**

00.JPG

* **on :** Elementos especificos en comun
* **inner :** Todo lo que hay en comun

```
# Ejemplo 01
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
```

```
df1
```

```
df2
```

```
px = pd.merge(df1,df2, left_on='lkey', right_on='rkey')
px 
```

```
df1.merge(df2, left_on='lkey', right_on='rkey')
```

```
df1.merge(df2, left_on='lkey', right_on='rkey',
          suffixes=('_left', '_right'))
```

```
# Ejemplo 02
df1 = pd.DataFrame({'a': ['foo', 'bar'], 'b': [1, 2]})
df2 = pd.DataFrame({'a': ['foo', 'baz'], 'c': [3, 4]})
```

```
df1
```

```
df2
```

```
df1.merge(df2, how='inner', on='a')
```

```
df1.merge(df2, how='left', on='a')
```

```
df1.merge(df2, how='right', on='a')
```

```
df1.merge(df2, how='outer')
```

```
dfx = pd.merge(df1,df2, how='outer')
dfx
```
